```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data = pd.read_csv('./finalapi.csv')
data.head()
```

```python
correlation_coeff, p_value = stats.pearsonr(data['WRTN_PREM_AMT'], data['PRD_INCRD_LOSSES_AMT'])

descriptive_stats = data.describe()

descriptive_stats
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='WRTN_PREM_AMT', y='PRD_INCRD_LOSSES_AMT')
plt.title('Written Premium vs Incurred Losses')
plt.xlabel('Written Premium Amount')
plt.ylabel('Incurred Losses Amount')
```

```python
print("\nCorrelation Analysis Results:")
print(f"Pearson correlation coefficient: {correlation_coeff:.4f}")
print(f"P-value: {p_value:.4f}")
print("\nDescriptive Statistics:")
print(descriptive_stats)

if p_value < 0.05:
    print("\nResult: Reject the null hypothesis")
    print("There is significant evidence of correlation between written premium and incurred losses")
else:
    print("\nResult: Fail to reject the null hypothesis")
    print("There is insufficient evidence of correlation between written premium and incurred losses")
```
